```python
from pathlib import Path
from datetime import datetime
from exif import Image, DATETIME_STR_FORMAT
```

Ce script concerne les fichiers mal nommés. Il est alors nécessaire qu'ils soient dans le répertoire `in`, dans un sous-répertoire du type `20240131blabla` (seule la partie `YYMMDD` est obligatoire). Chaque répertoire doit contenir au plus 59 photos. Les noms de fichiers et les metadonnées sont mis à jour en fonction du nom du répertoire et les fichiers traités sont placés dans le répertoire `out`.

```python
for dir in Path('in').iterdir():
    year, month, day = int(dir.name[:4]), int(dir.name[4:6]), int(dir.name[6:8])
    hour, minute = 23, 59
    print(year, month, day, hour, minute)
    second = 0
    for path in dir.iterdir():
        second += 1
        datetime_taken = datetime(year=year, month=month, day=day, hour=hour, minute=minute, second=second)
        new_file_name = datetime_taken.strftime('%Y%m%d_%H%M%S') + '_' + path.name
        image = Image(path)
        image.datetime_original = datetime_taken.strftime(DATETIME_STR_FORMAT)
        with open(f'out/{new_file_name}', 'wb') as new_image_file:
            new_image_file.write(image.get_file())
```
