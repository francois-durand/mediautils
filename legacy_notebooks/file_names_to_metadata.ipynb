```python
from pathlib import Path
from datetime import datetime
from exif import Image, DATETIME_STR_FORMAT
```

Ce script concerne les fichiers déjà bien nommés, par exemple `20240131_181253_blabla` (seule la partie `YYMMDD_HHMMSS` est obligatoire). Ces fichiers doivent être placés directement dans le répertoire `in`. Les metadonnées sont mises à jour en fonction du nom du fichier et les fichiers traités sont placés dans le répertoire `out`.

```python
for path in Path('in').iterdir():
    file_name = path.stem
    year, month, day = int(file_name[:4]), int(file_name[4:6]), int(file_name[6:8])
    hour, minute, second = int(file_name[9:11]), int(file_name[11:13]), int(file_name[13:15])
    datetime_taken = datetime(year=year, month=month, day=day, hour=hour, minute=minute, second=second)
    print(datetime_taken)
    image = Image(path)
    image.datetime_original = datetime_taken.strftime(DATETIME_STR_FORMAT)
    with open(f'out/{path.name}', 'wb') as new_image_file:
        new_image_file.write(image.get_file())
print('Done!')
```
