```python
from pathlib import Path
from datetime import datetime
from exif import Image, DATETIME_STR_FORMAT
from shutil import copy2
import os
import subprocess
```

```python
previous_date = ""
counter = 0
for path in Path('in').iterdir():
    file_name = path.stem
    date = file_name[4:12]
    year, month, day = int(date[0:4]), int(date[4:6]), int(date[6:8])
    if date != previous_date:
        previous_date = date
        counter = 1
    else:
        counter += 1
    second = counter % 60
    minute = 50 + (counter - second) // 60
    hour = 23
    datetime_taken = datetime(year=year, month=month, day=day, hour=hour, minute=minute, second=second)
    new_file_name = f"out/{year:02}{month:02}{day:02}_{hour:02}{minute:02}{second:02}_{path.name}"
    # print(datetime_taken)
    print(new_file_name)
    if path.suffix in {'.jpg', '.jpeg'}:
        image = Image(path)
        image.datetime_original = datetime_taken.strftime(DATETIME_STR_FORMAT)
        with open(new_file_name, 'wb') as new_image_file:
            new_image_file.write(image.get_file())
    elif path.suffix == '.mp4':
        # Copier le fichier vers le répertoire "out"
        copy2(path, new_file_name)
        # Mise à jour des timestamps (modification et dernier accès)
        mod_time = datetime_taken.timestamp()
        os.utime(new_file_name, (mod_time, mod_time))
        # Mise à jour des métadonnées avec ffmpeg
        # Format de date pour ffmpeg : "YYYY-MM-DD HH:MM:SS"
        metadata_date = datetime_taken.strftime('%Y-%m-%d %H:%M:%S')
        subprocess.run([
            "ffmpeg", "-i", new_file_name,
            "-metadata", f"creation_time={metadata_date}",
            "-codec", "copy", "-y", new_file_name
        ])
    else:
        raise ValueError("Unexpected extension")
print('Done!')
```
